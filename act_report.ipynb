## WERATEDOGS ARCHIVE TWEETS - INSIGHTS AND VISUALIZATION

The WeRateDogs tweets dataset contained names, dog stages, ratings as well as favorite count, retweets counts and dog breed prediction. As with any data analysis project, the goal is to gain insights by asking questions based on the data. For this project, descriptive and exploratory data analysis was performed to understand the data and find how different measures might be related to one another. From the analysis of the data, the following insights were obtained:

### 1. Most popular dog name


#### Figure 1: Most popular dog name in WeRateDogs tweets

The most popular dog name is Charlie, Cooper, Lucy, Oliver and Penny.

### 2. Most common dog stage


#### Figure 2: Most common dog stage among WeRateDogs tweets

It was observed that the most common dog stage was pupper with 212 original tweets. This is significantly higher than all the other dog stages. Doggo is the second most common with 74 tweets, followed by puppo with 24 tweets. The dog stage with the least number of tweets was floofer, with just 8 tweets.

### 3. Analysis of dog stages
The dog stages were subjected to further analysis to find out the favorite count and retweet count for each dog stage. In analyzing the dog stages based on favorite count and retweet count, we used the median as a more accurate measure of central tendency since both columns have extreme values and mean is easily affected by such extreme values. A plot of the favorite count distribution of each dog stage is included below.


#### Figure 3: Distribution of favorite count for each dog stage

From the above figure, there is no significant difference between the average number of likes (favorite count) among dog stages doggo and floofer. However, it is interesting to note that pupper, which is the most common dog stage in WeRateDogs tweets based on our previous finding, has the least number of likes as per median measure, while puppo, which is the second least common dog stage in WeRateDogs tweets, has the highest number of likes as per median measure.

A plot of the retweet count distribution of each dog stage was also made to provide more insights.


#### Figure 4: Distribution of retweet count for each dog stage

From the above figure, it can be observed that there is no significant difference between the average number of retweets among all the other dog stages except pupper, which has least number of retweets as per median measure.

### 4. Relationship between favorite count and retweet count



#### Figure 5: The relationship between favorite count and retweet count

From the above figure, there is a positive linear relationship between favorite count and retweet count. However, majority of the points are clustered at the beginning of the plot indicating that both variables have a positively skewed distribution. Hence, it is necessary to normalize the variables' values in order to investigate the strength of their relationship. Logarithmic transformation was used to normalise the values in figure 6 below.



#### Figure 6: Relationship after normalizing data

From the figure above, it is evident that there is a strong, positive linear relationship between favorite count and retweet count.

### 5. Most popular dog breeds 


#### Figure 7: Most popular dog breeds

The top ten most popular dog breeds are Golden Retriever, Labrador Retriever, Pembroke, Chihuahua, Pug, Chow, Samoyed, Pomeranian, Toy Poodle and Malamute.
